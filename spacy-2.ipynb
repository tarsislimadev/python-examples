```python
import en_core_web_sm
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
document = nlp('i bought a car')
document
```

```python
for token in document:
    print(token.text, token.lemma_, token.pos_)
```

```python
spacy.explain('VERB')
```

```python
for i in document:
    print(i.text, i.pos_, i.lemma_, i.tag_, i.dep_, i.shape, i.is_alpha, i.is_stop)
```

```python
spacy.explain('VBD')
```

```python
from spacy import displacy
displacy.render(document, style='dep', jupyter=True)
```
