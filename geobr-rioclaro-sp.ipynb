# Import libraries

```python
import geobr
```

# Load data

```python
muni = geobr.read_municipality(code_muni='SP', year=2024)
muni.head()
```

# Show and save state's image

```python
muni.plot(figsize=(16,9)).get_figure().savefig('sao_paulo.png')
```

# Show and save city's image

```python
muni_rc = muni[muni.name_muni == 'Rio Claro']
muni_rc.plot(figsize=(16,9)).get_figure().savefig('rio_claro.png')
```
