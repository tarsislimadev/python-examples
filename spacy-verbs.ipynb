```python
import pt_core_news_sm
```

```python
nlp = spacy.load('pt_core_news_sm')
```

```python
document = nlp('nós iremos comprar um carro')
document
```

```python
verbs = []
for token in document:
    if token.pos_ == 'VERB':
        verbs.append(token.text)
```

```python
count = len(verbs)
str(count) + ' verbo' + str('s' if count > 1 else '')
```

```python
' | '.join(verbs)
```
