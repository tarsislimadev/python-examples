```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
words = nlp('oi ola')
```

```python
word = ''
similarity = 1

while similarity > 0.3:
    word = input('digite uma palavra: ')
    similarity = words.similarity(nlp(word))
    print(f'similarity: {similarity}')
```
