```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
import random
```

```python
## palavras associadas a phishing
keywords = [
    'link',
    'prize',
    'reset',
    'verify',
    'reward',
    'urgent',
    'account',
    'password',
]
```

```python
## e-mails simulados
emails = [
  'Your account has been compromissed, please reset your password immediately',
  'Please, verify your identity by clicking the link below',
  'Your monthly report is ready for download',
  'You have won a prize! Click here to claim your reward'
]
email = random.choice(emails)
email
```

```python
dangerousness = 0
for keyword in keywords:
  dangerousness += nlp(keyword).similarity(nlp(email))
dangerousness /= len(keywords)
dangerousness
```

```python
if dangerousness > 0.25:
  print('Your mail messasge is dangerous!')
else:
  print('Your mail messasge is safe.')
```
