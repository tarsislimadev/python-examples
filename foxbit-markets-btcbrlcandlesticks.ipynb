```python
import requests as req
resp = req.get("https://api.foxbit.com.br/rest/v3/markets/btcbrl/candlesticks?interval=1d").json()
```

```python
from datetime import date
dates = [date.fromtimestamp(float(x[0]) // 1000) for x in resp]
prices = [float(x[1]) for x in resp]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(dates, prices)
plt.show()
```
