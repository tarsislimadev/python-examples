# Import libraries

```python
import matplotlib.pyplot as plt
import geobr
import geopandas as gpd
```

# Load data

```python
muni = geobr.read_municipality(code_muni='SP', year=2024)
muni.head()
```

# Show and save state's image

```python
fig, ax = plt.subplots(figsize=(9, 16))
muni.plot(figsize=(9, 16), ax=ax)
ax.axis('off')
fig.savefig('sao_paulo.png')
```

# Show and save city's image

```python
fig_rc, ax_rc = plt.subplots(figsize=(9, 16))
muni_rc = muni[muni.name_muni == 'Rio Claro']
muni_rc.plot(figsize=(9, 16), ax=ax_rc)
ax_rc.axis('off')
fig_rc.savefig('rio_claro.png')
```
