```python
import matplotlib.pyplot as plt
import requests as r
```

```python
symbol = "BTCBRL"
interval = "1d"
limit = 360
url = f"https://api2.binance.com/api/v3/klines?symbol={symbol}&interval={interval}&limit={limit}"
url
```

```python
resp = r.get(url).json()
resp
```

```python
import datetime
```

```python
dates = [datetime.datetime.fromtimestamp(r[0] / 1e3) for r in resp]
open_prices   = [round(float(r[1])) for r in resp]
high_prices   = [round(float(r[2])) for r in resp]
low_prices    = [round(float(r[3])) for r in resp]
close_prices  = [round(float(r[4])) for r in resp]

open_prices, high_prices, low_prices, close_prices
```

```python
fig, ax = plt.subplots(figsize=(16,9))

for i in range(len(dates)):
    # Choose color based on bullish/bearish
    color = 'green' if close_prices[i] >= open_prices[i] else 'red'
    
    # Draw the wick (high-low line)
    ax.vlines(i, low_prices[i], high_prices[i], color=color)
    
    # Draw the body (open-close rectangle)
    lower = min(open_prices[i], close_prices[i])
    height = abs(close_prices[i] - open_prices[i])
    ax.add_patch(plt.Rectangle((i-0.3, lower), 0.6, height, color=color))

ax.set_title(symbol)
plt.show()
```
