```python
import requests, matplotlib.pyplot as plt
```

```python
symbol = "BTCUSDT"
interval = "1m"
limit = 360
url = f"https://api2.binance.com/api/v3/klines?symbol={symbol}&interval={interval}&limit={limit}"
url
```

```python
resp = requests.get(url).json()
resp
```

```python
open_time = [r[0] for r in resp]
open_price = [float(r[1]) for r in resp]
open_time, open_price
```

```python
plt.figure(figsize=(12, 6))
plt.plot(open_time, open_price)
plt.show()
```
