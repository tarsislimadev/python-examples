```python
import argostranslate as argo
```

```python
from_code = "en"
to_code = "es"
```

# Download and install Argos Translate package

```python
argo.package.update_package_index()
```

```python
available_packages = argo.package.get_available_packages()
```

```python
package_to_install = next(
    filter(
        lambda x: x.from_code == from_code and x.to_code == to_code, available_packages
    )
)
```

```python
argo.package.install_from_path(package_to_install.download())
```

# Translate


```python
argo.translate.translate("Hello World", from_code, to_code)
```
