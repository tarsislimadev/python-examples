Import libraries

```python
import requests
import pandas as pd
import json
import time
```

Set up the request to Binance

```python
columns=["kline_open_time","open_price","high_price","low_price","close_price","volume","kline_close_time","quote_asset_volume","number_of_trades","taker_buy_base_asset_volume","taker_buy_quote_asset_volume","unused_field"]
def get_binance_data(symbol="BTCBRL", interval="1d", limit="1000"):
  time.sleep(1)
  url=f"https://api2.binance.com/api/v3/klines?symbol={symbol}&interval={interval}&limit={limit}"
  resp=requests.get(url)
  json_data=resp.json()
  df=pd.DataFrame(json_data, columns=columns)
  return df
```

Request Binance data

```python
get_binance_data()
```
