```python
document = nlp('i am learning natural language processing. we are in brazil.')
```

```python
for token in document:
  if nlp.vocab[token.text].is_stop:
    print(token.text)
```

```python
for token in document:
  if not nlp.vocab[token.text].is_stop:
    print(token.text)
```

```python
# exemplo 1
w1 = nlp('hello')
w2 = nlp('hi')
w3 = nlp('or')
```

```python
w1.similarity(w3)
```

```python
text1 = nlp('when will the new movie be released?')
text2 = nlp('the new movie will be released next month')
text3 = nlp('what color is the car?')
```

```python
text1.similarity(text2)
```

```python
# exemplo 2
text = nlp('cat dog horse person')
```

```python
for text1 in text:
  print(2*'-', text1.text, 2*'-')
  for text2 in text:
    similarity = text1.similarity(text2) * 100
    print(f'{similarity:.4f}% between {text1} and {text2}')
```
