```python
import argostranslate.package
import argostranslate.translate
```

```python
import requests as r
```

```python
from_code = "pt"
to_code = "es"
```

# Download and install Argos Translate package


```python
argostranslate.package.update_package_index()
```

```python
available_packages = argostranslate.package.get_available_packages()
```

```python
package_to_install = next(
    filter(
        lambda x: x.from_code == from_code and x.to_code == to_code, available_packages
    )
)
```

```python
argostranslate.package.install_from_path(package_to_install.download())
```

# Get text

```python
datetime = "20260226"
```

```python
resp = r.get(f"https://www.casadoscontos.com.br/texto/{datetime}").text
```

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, "html.parser")
soup
```

```python
text = soup.find_all('article')[0].text
text
```

# Translate


```python
import re
paragraphs = [t.strip().lower() for t in re.split(r'[.!?]', text) if len(t) > 0]
```

```python
translated_texts = []
for text in paragraphs:
  translated_text = argostranslate.translate.translate(text, from_code, to_code)
  translated_texts.append(translated_text)
  print(translated_text)
```

```python
with open('translated-casadoscontos.txt', 'a+') as file:
  for t in translated_texts:
    file.write(f"{t} \n")
```
