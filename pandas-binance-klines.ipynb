```python
import requests
import pandas as pd
```

```python
url="https://api2.binance.com/api/v3/klines?symbol=BTCBRL&interval=1d&limit=400"
```

```python
resp=requests.get(url)
resp_json=resp.json()
```

```python
klines=pd.DataFrame(resp_json)
klines
```
